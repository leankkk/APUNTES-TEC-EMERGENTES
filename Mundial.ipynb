```python
# Por cada partido, cual fue el equipo que gano?
# Completar
# Si el equipo local hizo mas goles, gana el local
datos.loc[datos["1_goals"] > datos["2_goals"], "ganador"] = datos["1"]
# Si el equipo visitante hizo mas goles, gana el visitante
datos.loc[datos["2_goals"] > datos["1_goals"], "ganador"] = datos["2"]

print(datos[["1", "2", "1_goals", "2_goals", "ganador"]])
```

```python
# En cada fase del mundial, cual fue el equipo que hizo mas goles?
# agarra los goles de cada equipo cuando juega de local
goles_1 = datos[["group", "1", "1_goals"]].rename(columns={"1": "equipo", "1_goals": "goles"})
# agarramos los goles de cada equipo cuando juega de visitante
goles_2 = datos[["group", "2", "2_goals"]].rename(columns={"2": "equipo", "2_goals": "goles"})
#unir
todos = pd.concat([goles_1, goles_2])

# unifica grupos en una sola fase
todos.loc[todos["group"].str.contains("Group"), "group"] = "Fase de Grupos"

por_fase = todos.groupby(["group", "equipo"])["goles"].sum()
print(por_fase.groupby("group").idxmax())


#  en la final y cuartos da argentina, porque es el primero en orden alfabetico. 
```
