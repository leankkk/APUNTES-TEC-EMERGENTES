```python
import pandas as pd

df = pd.read_csv('qatar.csv')

total_partidos = len(df)
print("Total de partidos jugados: ", + total_partidos)
```

```python
import pandas as pd
todas = pd.concat([df["1"], df["2"]]).unique()
total = len(todas)
print("Total de selecciones participantes: ", total)
```

```python
fases = df["group"].unique()
print("Total de fases: " + str(len(fases )-8 ))
```

```python
promedio = df["attendance"].mean()
print("Attendance promedio por partido: " + str(promedio))
```

```python
df["diferencia"] = abs(df["1_goals"] - df["2_goals"])
maximo = df["diferencia"].max()
partido = df[df["diferencia"] == maximo]
print(partido[["1", "2", "1_goals", "2_goals", "diferencia"]])
```

```python
penales= datos.groupby("1")["1_panelties_scored"].sum() +  datos.groupby("2")["2_panelties_scored"].sum()
print(penales.idxmax())
```
